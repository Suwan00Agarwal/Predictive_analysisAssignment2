```python
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit
```

```python
roll_no = 102317217
a_r = 0.05 * (roll_no % 7)
b_r = 0.3 * (roll_no % 5 + 1)
print(f"Constants calculated: a = {a_r:.4f}, b = {b_r:.4f}")
```

```python
try:
    data = pd.read_csv('data.csv', encoding='latin1', low_memory=False)
except UnicodeDecodeError:
    data = pd.read_csv('data.csv', encoding='ISO-8859-1', low_memory=False)
```

```python
no2_vals = data['no2'].dropna().values
z_vals = no2_vals + a_r * np.sin(b_r * no2_vals)
```

```python
def gaussian_pdf(z, scale, lam, center):
    return scale * np.exp(-lam * (z - center) ** 2)
```

```python
hist_vals, edges = np.histogram(z_vals, bins=100, density=True)
z_mid = (edges[:-1] + edges[1:]) / 2
```

```python
start_params = [np.max(hist_vals), 0.1, np.mean(z_vals)]

opt_params, cov_matrix = curve_fit(
    gaussian_pdf,
    z_mid,
    hist_vals,
    p0=start_params
)
```

```python
c_est, lambda_est, mu_est = opt_params
print("\nðŸ”¹ Learned Parameters for Submission:")
print(f"-> c      : {c_est:.6f}")
print(f"-> lambda : {lambda_est:.6f}")
print(f"-> mu     : {mu_est:.6f}")
```
